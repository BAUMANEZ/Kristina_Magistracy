```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import make_blobs 
np.random.seed(0)
n_samples = 100
random_state = 3
```

```python
# Вариант 1
centers = [(0.2, 3), (1, 3)]
blobs = make_blobs(n_samples=n_samples,
                             centers=centers,
                             cluster_std=[(0.07, 0.5), (0.06, 0.5)],
                             random_state=random_state)
bl = blobs[0] 
plt.scatter(bl[:, 0], bl[:, 1], marker='o')
```
